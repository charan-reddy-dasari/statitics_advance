1. What is a random variable in probability theory?
  - A random variable is a function that assigns numerical values to outcomes of a random experiment. It can be either discrete or continuous.

2. What are the types of random variables?

  - Discrete Random Variable: Takes a finite or countably infinite set of values (e.g., number of heads in coin tosses).
  - Continuous Random Variable: Takes an infinite number of values within a given range (e.g., height of students).
3. What is the difference between discrete and continuous distributions?

  - Discrete Distribution: Deals with distinct values (e.g., Binomial, Poisson).
  - Continuous Distribution: Deals with an interval of values (e.g., Normal, Exponential).
4. What are probability distribution functions (PDF)?
  - A probability distribution function describes the likelihood of different outcomes in a random variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

  - PDF gives the probability density at a specific point.
  - CDF gives the probability of a variable being less than or equal to a specific value.
6. What is a discrete uniform distribution?
  - A distribution where all possible outcomes have equal probability (e.g., rolling a fair die).

7. What are the key properties of a Bernoulli distribution?

  - Two possible outcomes: success (1) or failure (0).
  - Probability of success: p
  - probability of failure: 1-p
  - Mean: p
  - variance = p(1-p)
8. What is the binomial distribution, and how is it used in probability?
  - The binomial distribution models the number of successes in n
  - n independent Bernoulli trials with success probability p.

9. What is the Poisson distribution and where is it applied?
  - A discrete probability distribution modeling the number of events in a fixed interval, given an average rate λ. Used in event occurrence modeling (e.g., call center arrivals).

10. What is a continuous uniform distribution?
  - A probability distribution where all outcomes in a continuous range are equally likely.

11. What are the characteristics of a normal distribution?

  - Bell-shaped curve.
  - Symmetrical around the mean.
  - Defined by mean 𝜇
  - standard deviation σ.
  - 68%-95%-99.7% rule applies.
12. What is the standard normal distribution, and why is it important?
  - A normal distribution with mean = 0 and standard deviation = 1. It simplifies statistical calculations and comparisons.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - CLT states that the sampling distribution of the sample mean approaches a normal distribution, regardless of the population distribution, as sample size increases.

14. How does the Central Limit Theorem relate to the normal distribution?
  - CLT ensures that sample means follow a normal distribution, allowing statistical inference.

15. What is the application of Z statistics in hypothesis testing?
  - Z-statistics are used when sample size is large and population variance is known. It helps determine whether to reject the null hypothesis.

16. How do you calculate a Z-score, and what does it represent?

  - Z = (Data Point - Mean)/Standard Deviation

  - It represents how many standard deviations a value is from the mean.

17. What are point estimates and interval estimates in statistics?

  - Point Estimate: A single value estimate of a population parameter (e.g., sample mean).
  - Interval Estimate: A range of values within which a parameter likely falls (e.g., confidence interval).
18. What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals provide a range within which the true parameter value is expected to lie with a given probability.

19. What is the relationship between a Z-score and a confidence interval?
  - Z-scores determine the critical values used to construct confidence intervals.

20. How are Z-scores used to compare different distributions?
  - Z-scores standardize values, enabling comparison across different distributions.

21. What are the assumptions for applying the Central Limit Theorem?

  - Random sampling
  - Independent observations
  - Sufficiently large sample size (typically n>30)
22. What is the concept of expected value in a probability distribution?
The expected value is the long-run average of a random variable’s outcomes, given by: E(X)=∑xP(x)(discrete),E(X)=∫xf(x)dx(continuous)
23. How does a probability distribution relate to the expected outcome of a random variable?
- The expected value provides the central tendency of the distribution, predicting long-run behavior.

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
from scipy.stats import norm
data = np.random.normal(0, 1, 100)
z_score = (data[0] - np.mean(data)) / np.std(data)
print("Z-score:", z_score)
print("Probability:", norm.cdf(z_score))
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm
sample = np.random.normal(50, 10, 30)
pop_mean = 50
z_stat = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats
data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std_err = stats.sem(data)
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)
print("Confidence Interval:", conf_int)
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats
data = np.random.normal(100, 15, 50)
conf_int = stats.norm.interval(0.95, loc=np.mean(data), scale=stats.sem(data))
print("Confidence Interval:", conf_int)
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np
data = np.random.normal(50, 10, 100)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
print("Range:", np.max(data) - np.min(data))
print("IQR:", iqr)
```
